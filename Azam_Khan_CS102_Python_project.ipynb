Importing the necessary libraries and tools

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

Creating dataframes for the CSV files and selecting the relevant years and data from each CSV

Note: This part proved unexpectedly problematic, since Panda was reading the values of the CSV as strings, and it was unclear why it was not plotting the graphs right. You can see at the end of each block a line to convert the value type to float(and for sales values, removing commas to make this possible).


```python
fuel_name = ["Fuel Usage"]
data_source1 = 'Fuel_usage.csv'
df_fuel = pd.read_csv(data_source1, usecols=['Value'])
df_fuel_reduced = df_fuel.iloc[112:142, 0]
df_fuel_reduced = df_fuel_reduced.astype(np.float)

data_source2 = 'Truck_sales.csv'
df_sales = pd.read_csv(data_source2)
df_sales_reduced = df_sales.iloc[4, 1:31]
df_sales_reduced = df_sales_reduced.str.replace(',','').astype(np.float)
```

Creating a steady counter for the years 1990-2019 for the x axis

```python
years = np.arange(1990,2020,1)
```

Plotting the data onto the same plot using the matplotlib subplots command

```python
fig, sales = plt.subplots()
sales.set_xlabel('Year')

sales.plot(years,df_sales_reduced.values,scaley=True, color='red')
sales.set_ylabel('Light truck sales (Units)', color='red')
sales.tick_params(axis='y', labelcolor='red')

fuel = sales.twinx()
fuel.plot(years, df_fuel_reduced.values, scaley=True, color='blue', )
fuel.set_ylabel('Annual fuel consumption per vehicle (Gallons)',color='blue')
fuel.tick_params(axis='y', labelcolor='blue')

plt.show
```
